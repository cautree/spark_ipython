```python
#Lazy evaluation
arr = [rd.gauss(5,10) for x in xrange(10000)]
```

```python
# Calculate the average
rdd_arr = sc.parallelize(arr)
rdd_arr.cache()
%timeit sum_rdd_arr = rdd_arr.reduce(lambda x,y:x+y) 
%timeit avg_rdd_arr = sum_rdd_arr / rdd_arr.count()
print sum_rdd_arr
print avg_rdd_arr
```

```python
# Calculate the standard deviation
rdd_t1 = rdd_arr.map(lambda x: x-avg_rdd_arr)
rdd_t2 = rdd_t1.map(lambda x:x**2)
print rdd_t1
print rdd_t2
```

```python
# It does't run until a reduce command
sum_t3 = rdd_t2.reduce(lambda x,y:x+y) / rdd_arr.count()
std_rdd_arr = np.sqrt(sum_t3)
print std_rdd_arr
```

```python
print 'numpy: ',np.sum(arr),np.mean(arr), np.std(arr)
print 'spark:', sum_rdd_arr, avg_rdd_arr, std_rdd_arr
```

```python
data = 'Logic will get you from A to B. Imagination will take you everywhere'

## Split the setence in words
words = str(data).split(' ')
```

```python
## Build a RDD and parallelize it on the cluster
dataRDD = sc.parallelize(words)
dataRDD.cache()
```

```python
## Transform: lower case all elements in dataRDD.
## Spark will work only after an action command 
dataRDD_lower = dataRDD.map(lambda x:x.lower())
dataRDD_lower
```

```python
## Now Spark will run the transformation and the action
## Collect() will display all elements in RDD
dataRDD_lower.collect()
```
